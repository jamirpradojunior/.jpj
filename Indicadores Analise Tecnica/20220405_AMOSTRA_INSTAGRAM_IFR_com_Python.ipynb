# **3. Indicadores de Análise Técnica**

### **3.2. Osciladores**


#### 3.2.1. Índice de Força Relativa (IFR)


#### *Briefing IFR:*
Tem o objetivo de comparar a "força" entre compradores e vendedores de um ativo naquele momento. O IFR varia entre 0 e 100, onde valores próximos a 0 indicam força vendedora dominante e, proximos a 100, compradora dominante. O IFR é calculado usando a seguinte fórmula:<p>
$ IFR = 100 - {\left(\frac{100}{1 + {\left(\frac{A}{B}\right)}}\right)} $

Onde:<p>
*A* = média dos preços dos últimos períodos de alta (usualmente fechamento e 14 períodos);<p>
*B* = média dos preços dos últimos períodos de baixa.<p>

#### Cálculo IFR

```python
rsi = ta.momentum.RSIIndicator(close=data_ohlcv['Adj Close'],window=14)
data_ohlcv['IFR_14p'] = rsi.rsi()
```

#### Visualização IFR

```python
#data_range_graph = pd.date_range(start='2021-01-01',end='2021-06-30')
#data_reduzido = data_ohlcv[data_ohlcv.index.isin(data_range_graph)]

fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Candlestick(name='PETR4', x=data_ohlcv.index, open=data_ohlcv['Open'], high = data_ohlcv['High'], low=data_ohlcv['Low'], close=data_ohlcv['Close']), row=1,col=1)
fig.add_trace(go.Scatter(name='IFR_14p', x=data_ohlcv.index, y=data_ohlcv['IFR_14p'],marker_color='blue'), row=2, col=1)
fig.add_hline(y=70, line_width=3, line_dash="dash", line_color="black",row=2, col=1)
fig.add_hline(y=30, line_width=3, line_dash="dash", line_color="black",row=2, col=1)

fig.update_xaxes(rangebreaks=[
    dict(bounds=["sat", "mon"]), # não mostrar finais de semana
    dict(bounds=[18, 10], pattern="hour"), # não mostrar horas de pregao fechado (observar cada ativo específico)
    ])

fig.update_layout(xaxis_rangeslider_visible=False, title_text='Bandas de Bollinger',template = 'simple_white',width=500,height=500)
fig.show()
```
