## **3.1.1. Médias Móveis (SMA e EMA)**


#### *Briefing:*

Estão entre os grandes pilares da Análise Técnica. São calculadas a partir de amostras sequenciais de séries temporais (no caso, os últimos '*n*' preços do ativo) em um determinado timeframe. Por exemplo, uma média móvel (MM) aritmética de 9 períodos representa a média dos últimos 9 dias de pregão em um timeframe diário. Principais objetivos são suavizar flutuações curtas e destacar tendências de maior prazo. Criticadas devido ao atraso (*Lag Factor*) na definição de uma tendencia, apresentam problemas em mercados lateralizados ou quando o preço inicia um movimento com grande aceleração. Existem diversos tipos de MMs: aritmética (ou simples), exponencial, ponderada, Welles-Wilder, ...

## **3.1.2. Moving Average Convergence/Divergence (MACD)**


#### *Briefing:*

Como o nome diz, busca explorar a diferença entre duas médias móveis (MM_longa e MM_curta). Usualmente, são MM exponenciais de 26 e 12 períodos. Quando observamos sua estrutura, muitas premissas estão baseadas na estratégia de cruzamentos de médias. Por ex., quando a MM12 está acima da MM26 e se distanciando, temos um potencial contexto de alta em aceleração.

#### *Cálculos:*

Linha MACD:

A “distância” (MM_curta – MM_longa) entre as duas MM, leva ao primeiro parâmetro do indicador: a linha MACD

```python
ema_long = ta.trend.EMAIndicator(data_ohlcv[('Close','ITUB4')], window=26)
ema_short = ta.trend.EMAIndicator(data_ohlcv[('Close','ITUB4')], window=12)
data_ohlcv[('MM','EMA_long')] = ema_long.ema_indicator()
data_ohlcv[('MM','EMA_short')] = ema_short.ema_indicator()
data_ohlcv.head(30)
```

# **3.2. Osciladores**


## 3.2.1. Índice de Força Relativa (IFR)


### *Briefing IFR:*
Tem o objetivo de comparar a "força" entre compradores e vendedores de um ativo naquele momento. O IFR varia entre 0 e 100, onde valores próximos a 0 indicam força vendedora dominante e, proximos a 100, compradora dominante. O IFR é calculado usando a seguinte fórmula:<p>
$ IFR = 100 - {\left(\frac{100}{1 + {\left(\frac{A}{B}\right)}}\right)} $

Onde:<p>
*A* = média dos preços dos últimos períodos de alta (usualmente fechamento e 14 períodos);<p>
*B* = média dos preços dos últimos períodos de baixa.<p>

# **3.4. Indicadores de volatilidade**


Volatitidade pode ser definida como a capacidade de sofrer mudanças constantemente. No mercado financeiro, é uma medida de dispersão dos preços de um ativo em relação a algum ponto de referência (ex., à própria MM ou a um ETF). Também é associada a uma "medida de risco", onde quanto maior a volatilidade (mais o preço de um ativo variar em um determinado período de tempo), maior o risco de se ganhar ou perder dinheiro nas negociações.<p>
No caso da maioria dos indicadores de volatilidade, os próprios preços do ativo (MMs e OHLC) são utilizados nos cálculos.

### **3.4.1. Bandas de Bollinger**

#### *Briefing:*

Apesar do preço de um ativo variar bastante acima e abaixo das MMs, ele dificilmente foge de uma determinada região. Nesse caso, o preço acaba sendo constantemente atraído para uma zona de equilíbrio (próximo das MMs). Assim, o indicador Bandas de Bollinger consiste de duas linhas (superior e inferior) traçadas a partir de uma determinada distância de uma MM. A distância da MM é baseada no desvio-padrão (DP) dos últimos períodos determinados pela MM. Usualmente utiliza-se DP = 2.  


#### *Cálculos:*

$$
\mathbf{BB_{med} = \text{MM}_{n}\\
BB_{superior} = \text{MM}_{n} + k \times {SD}_{n} \\ 
BB_{inferior} = \text{MM}_{n} - k \times {SD}_{n}}
$$

onde:<p>
BB = banda de Bollinger;<p>
MM<sub>n</sub> = Média Móvel dos últimos *n* períodos da série temporal;<p>
k = fator de multiplicação do desvio-padrão;<p>
SD = desvio-padrão dos preços (usualmente fechamento) dos últimos *n* périodos da série temporal;


```python
bb = ta.volatility.BollingerBands(close=data_ohlcv[('Close','ITUB4')], window=20, window_dev=2)
data_ohlcv[('Volatil','BB_med')] = bb.bollinger_mavg()
data_ohlcv[('Volatil','BB_sup')] = bb.bollinger_hband()
data_ohlcv[('Volatil','BB_inf')] = bb.bollinger_lband()
```

```python
## Vizualização Bandas de Bollinger

data_range_graph = pd.date_range(start='2021-01-01',end='2021-06-30')
data_reduzido = data_ohlcv[data_ohlcv.index.isin(data_range_graph)]

fig = make_subplots(rows=1, cols=1)

fig.add_trace(go.Candlestick(name='ITUB4', x=data_reduzido.index, open=data_reduzido[('Open','ITUB4')], high = data_reduzido[('High','ITUB4')], low=data_reduzido[('Low','ITUB4')], close=data_reduzido[('Close','ITUB4')]), row=1,col=1)
fig.add_trace(go.Scatter(name='BB_med', x=data_reduzido.index, y=data_reduzido[('Volatil','BB_med')],marker_color='blue'), row=1, col=1)
fig.add_trace(go.Scatter(name='BB_sup', x=data_reduzido.index, y=data_reduzido[('Volatil','BB_sup')],marker_color='orange'), row=1, col=1)
fig.add_trace(go.Scatter(name='BB_inf', x=data_reduzido.index, y=data_reduzido[('Volatil','BB_inf')],marker_color='orange'), row=1, col=1)

fig.update_layout(xaxis_rangeslider_visible=False, title_text='Bandas de Bollinger',template = 'simple_white',width=500,height=500)
fig.show()
```

### **3.4.2. True Range (TR)**

#### *Briefing:*

Em uma tradução literal, True Range (TR) significa algo como amplitude verdadeira ou real. Embora raramente utilizado como um indicador propriamente dito, o TR é a base para outros indicadores e parâmetros da análise técnica, como o Stop ATR (Average True Range) e o OBTR (On-Balance True Range). O TR busca avaliar a amplitude atingida pelos preços de um ativo em determinado período de forma mais "confiável". Para isso, ele considera não apenas a tradicional amplitude (máx - mín) de preços, mas também o preço de fechamento anterior (incluing potenciais gaps).
\
\
O TR é representado pelo maior valor (absoluto, desconsiderando o sinal) calculado dentre as três condições:
\
\
$
\mathrm{\text{1. Maximo}_{atual} - {Minimo}_{atual}\\
\text{2. Maximo}_{atual} - {Fechamento}_{anterior}\\
\text{3. Minimo}_{atual} - {Fechamento}_{anterior}\\}
$

<img src="https://school.stockcharts.com/lib/exe/fetch.php?media=technical_indicators:average_true_range_atr:atr-1-trexam.png" width="350" height="250">

```python
## Representação gráfica TR

fig = make_subplots(rows=1, cols=1)

fig.add_trace(go.Candlestick(name='ITUB4', x=data_reduzido.index, open=data_reduzido[('Open','ITUB4')], high = data_reduzido[('High','ITUB4')], low=data_reduzido[('Low','ITUB4')], close=data_reduzido[('Close','ITUB4')]), row=1,col=1)
```

#### *Cálculos:*

$$
\mathbf{BB_{med} = \text{MM}_{n}\\
BB_{superior} = \text{MM}_{n} + k \times {SD}_{n} \\ 
BB_{inferior} = \text{MM}_{n} - k \times {SD}_{n}}
$$

onde:<p>
BB = banda de Bollinger;<p>
MM<sub>n</sub> = Média Móvel dos últimos *n* períodos da série temporal;<p>
k = fator de multiplicação do desvio-padrão;<p>
SD = desvio-padrão dos preços (usualmente fechamento) dos últimos *n* périodos da série temporal;


```python
bb = ta.volatility.BollingerBands(close=data_ohlcv[('Close','ITUB4')], window=20, window_dev=2)
data_ohlcv[('Volatil','BB_med')] = bb.bollinger_mavg()
data_ohlcv[('Volatil','BB_sup')] = bb.bollinger_hband()
data_ohlcv[('Volatil','BB_inf')] = bb.bollinger_lband()
```

## **3.5.1 On-balance Volume (OBV)**

#### *Briefing:*

Em uma tradução literal, o OBV seria "volume em equilíbrio". E seu objetivo é exatamente esse, avaliar como as movimentações dos preços estão em equilíbrio com o volume. O OBV é um bom detector de quão consistente foi a movimentação dos preços. Quando consistente, o gráfico OBV resulta em picos e vales similares aos observados na movimentação dos preços. Quando há divergências, o OBV pode estar sinalizando um final de tendência ou até forte reversão dos preços. Por isso, alguns dizem que ele pode ser usado como termômetro para a "euforia do mercado".

#### *Cálculos:*

$$
\mathbf{OBV = {OBV}_{anterior} + \begin{Bmatrix}
volume & \ \ se \ \ Fech > {Fech}_{anterior} \\
0 & \ \ se \ \ Fech = {Fech}_{anterior} \\
volume & \ \ se \ \ Fech <  {Fech}_{anterior} \\
\end{Bmatrix}}
$$

```python
obv = ta.volume.OnBalanceVolumeIndicator(data_ohlcv[('Close','ITUB4')], data_ohlcv[('Volume','ITUB4')])
data_ohlcv[('Volume_Ind','OBV')] = obv.on_balance_volume()
data_ohlcv.tail(50)
```

## **3.5.2 Volume Weighted Average Price (VWAP)**

#### *Briefing:*

A média dos tubarões

#### *Cálculos:*

$$
\mathbf{VWAP = \left ( \frac{\sum \left ( {Preco}_{i} \times {Qtd}_{i} \right )}{\sum {Qtd}_{j}} \right )}
$$

onde:
VWAP = preço médio ponderado pelo volume;<p>
Preco<sub>i</sub> = Preço do ativo (usualmente fechamento) no *i* período da série temporal;<p>
Qtd<sub>i</sub> = Volume em quantidade negociado no *i* período da série temporal;<p>
Qtd<sub>j</sub> = Volume em quantidade negociado ao longo de todo o período *j* avaliado na série temporal;


```python
vwap = ta.volume.VolumeWeightedAveragePrice(data_ohlcv[('High','ITUB4')],data_ohlcv[('Low','ITUB4')], data_ohlcv[('Close','ITUB4')], data_ohlcv[('Volume','ITUB4')],window=20)
data_ohlcv[('Volume_Ind','VWAP')] = vwap.volume_weighted_average_price()
```
