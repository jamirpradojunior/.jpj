# **1. EDIÇÃO 1 - BITCOIN**

```python
!pip install yfinance -- quiet
```

```python
import pandas as pd
import yfinance as yf
```

```python
lista_ativos = ['BRL=X','BTC-USD']
```

```python
OHLC_ativos = yf.download(lista_ativos, period='max', interval = "1D")['Adj Close']
OHLC_ativos
```

## **1.1. Há exatos 7 anos**

```python
OHLC_ativos_2015 = OHLC_ativos.loc['2015-04-21':]
OHLC_ativos_2015
```

```python
OHLC_ativos_2015['BTC-BRL'] = OHLC_ativos_2015['BTC-USD'] * OHLC_ativos_2015['BRL=X']
OHLC_ativos_2015
```

```python
OHLC_ativos_2015['BRL_100_qts_BTC'] = 100/OHLC_ativos_2015['BTC-BRL']
OHLC_ativos_2015
```

```python
OHLC_ativos_2015['Hoje'] = OHLC_ativos_2015['BRL_100_qts_BTC'].iloc[0]*OHLC_ativos_2015['BTC-BRL']
OHLC_ativos_2015
```

## **1.2. Há exatos 5 anos**

```python
OHLC_ativos_2017 = OHLC_ativos.loc['2017-04-21':]
OHLC_ativos_2017['BTC-BRL'] = OHLC_ativos_2017['BTC-USD'] * OHLC_ativos_2017['BRL=X']
OHLC_ativos_2017['BRL_100_qts_BTC'] = 100/OHLC_ativos_2017['BTC-BRL']
OHLC_ativos_2017['Hoje'] = OHLC_ativos_2017['BRL_100_qts_BTC'].iloc[0]*OHLC_ativos_2017['BTC-BRL']
OHLC_ativos_2017
```

## **1.3. Há exatos 3 anos**

```python
OHLC_ativos_2019 = OHLC_ativos.loc['2019-04-19':]
OHLC_ativos_2019['BTC-BRL'] = OHLC_ativos_2019['BTC-USD'] * OHLC_ativos_2019['BRL=X']
OHLC_ativos_2019['BRL_100_qts_BTC'] = 100/OHLC_ativos_2019['BTC-BRL']
OHLC_ativos_2019['Hoje'] = OHLC_ativos_2019['BRL_100_qts_BTC'].iloc[0]*OHLC_ativos_2019['BTC-BRL']
OHLC_ativos_2019
```

## **1.4. Há exato 1 ano**

```python
OHLC_ativos_2021 = OHLC_ativos.loc['2021-04-21':]
OHLC_ativos_2021['BTC-BRL'] = OHLC_ativos_2021['BTC-USD'] * OHLC_ativos_2021['BRL=X']
OHLC_ativos_2021['BRL_100_qts_BTC'] = 100/OHLC_ativos_2021['BTC-BRL']
OHLC_ativos_2021['Hoje'] = OHLC_ativos_2021['BRL_100_qts_BTC'].iloc[0]*OHLC_ativos_2021['BTC-BRL']
OHLC_ativos_2021
```

## **1.5. Gráfico normalizado**

```python
import matplotlib.pyplot as plt

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
OHLC_ativos = yf.download(lista_ativos, start = '2015-04-21', interval = "1d")['Adj Close']
OHLC_ativos
```

```python
OHLC_ativos = yf.download(lista_ativos, start = '2015-04-21', interval = "1wk")['Adj Close']
OHLC_ativos
```

```python
OHLC_ativos_2015 = OHLC_ativos.loc['2015-04-21':]
OHLC_ativos_2015['Data'] = OHLC_ativos_2015.index
OHLC_ativos_2015['Bitcoin_BRL'] = OHLC_ativos_2015['BRL=X']*OHLC_ativos_2015['BTC-USD']
OHLC_ativos_2015
```

```python
fig = px.line(OHLC_ativos_2015, x="Data", y="Bitcoin_BRL")


fig.update_layout(title='<b>Bitcoin BRL ao longo dos últimos 7 anos', xaxis_title="<b>Data", yaxis_title="<b>Bitcoin (R$)", legend_title="",font=dict(family="Arial, Arial, Arial",size=15,color="black"))
fig.update_layout(xaxis_rangeslider_visible=False, margin=dict(l=100, r=20, t=80, b=20), template = 'simple_white', paper_bgcolor="#f0f5ff",width=800,height=500,showlegend=False)
fig.show()
```
